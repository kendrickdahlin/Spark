```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample data
processors = np.array([1, 8, 14, 28, 56])

data = {}

# Read CSV files
for i in processors:
    file_path = f"results_{i}.csv"
    data[i] = pd.read_csv(file_path).iloc[:, -1].to_numpy()

speedups = []

# Calculate speedup for each data size
for i in range(0, 11):
    index = 1000 + i * 100
    l = data[1][i] / np.array([data[1][i], data[8][i], data[14][i], data[28][i], data[56][i]])
    speedups.append((index, l))

# Plotting the graph
plt.figure(figsize=(12, 6))

for idx, speedup in speedups:
    plt.plot(processors, speedup, marker='o', linestyle='-', label=f'Data size {idx}')

plt.xlabel('Number of Processors')
plt.ylabel('Speedup')
plt.title('Speedup vs Number of Processors for Different Data Sizes')
plt.grid(True)
plt.xticks(processors)
plt.legend(title='Data Sizes')
plt.show()

scaleups = []

for i in processors:
    s = data[i] / data[i][0]  # Assuming data size increases linearly
    scaleups.append((i, s))

# Plotting the scale-up graph
plt.figure(figsize=(12, 6))

for proc, scaleup in scaleups:
    plt.plot(range(1000, 2100, 100), scaleup, marker='o', linestyle='-', label=f'{proc} Processors')

plt.xlabel('Data Size')
plt.ylabel('Scale-Up')
plt.title('Scale-Up vs Data Size for Different Numbers of Processors')
plt.grid(True)
plt.legend(title='Number of Processors')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample data
processors = np.array([1, 8, 14, 28, 56])

data = {}

# Read CSV files
for i in processors:
    file_path = f"results_{i}.csv"
    data[i] = pd.read_csv(file_path).iloc[:, -1].to_numpy()

speedups = []

# Calculate speedup for each data size
for i in range(0, len(data[1])):
    index = 1000 + i * 100
    l = data[1][i] / np.array([data[1][i], data[8][i], data[14][i], data[28][i], data[56][i]])
    speedups.append((index, l))

# Plotting the speedup graph
plt.figure(figsize=(12, 6))

for idx, speedup in speedups:
    plt.plot(processors, speedup, marker='o', linestyle='-', label=f'Data size {idx}')

plt.xlabel('Number of Processors')
plt.ylabel('Speedup')
plt.title('Speedup vs Number of Processors for Different Data Sizes')
plt.grid(True)
plt.xticks(processors)
plt.legend(title='Data Sizes')
plt.show()


scaleup = []
c=0
# Calculate scale-up for each number of processors
for i in processors:

    scaleup.append( data[1][0] / data[i][c])
    c=c+1

# Plotting the scale-up graph
plt.figure(figsize=(12, 6))

plt.plot(range(1000, 1000 + 100 * len(scaleup), 100), scaleup, marker='o', linestyle='-', label=f'{proc} Processors')

plt.xlabel('Data Size')
plt.ylabel('Scale-Up')
plt.title('Scale-Up vs Data Size for Different Numbers of Processors')
plt.grid(True)
plt.legend(title='Number of Processors')
plt.show()
```
